```python
yesbank["20d"] = np.round(yesbank["Adj Close"].rolling(window = 20, center = False).mean(), 2)
pandas_candlestick_ohlc(yesbank.loc['2019-01-04':'2019-12-31',:], otherseries = "20d")
```

```python
yesbank['20d-50d'] = yesbank['20d'] - yesbank['50d']
yesbank.tail()
```

```python
# np.where() is a vectorized if-else function, where a condition is checked for each component of a vector, and the first argument passed is used when the condition holds, and the other passed if it does not
yesbank["Regime"] = np.where(yesbank['20d-50d'] > 0, 1, 0)
# We have 1's for bullish regimes and 0's for everything else. Below I replace bearish regimes's values with -1, and to maintain the rest of the vector, the second argument is apple["Regime"]
yesbank["Regime"] = np.where(yesbank['20d-50d'] < 0, -1, yesbank["Regime"])
yesbank.loc['2016-01-04':'2019-12-31',"Regime"].plot(ylim = (-2,2)).axhline(y = 0, color = "black", lw = 2)
```

```python
yesbank["Regime"].value_counts()
```

```python
# To ensure that all trades close out, I temporarily change the regime of the last row to 0
regime_orig = yesbank.loc[:, "Regime"].iloc[-1]
yesbank.loc[:, "Regime"].iloc[-1] = 0
yesbank["Signal"] = np.sign(yesbank["Regime"] - yesbank["Regime"].shift(1))
# Restore original regime data
yesbank.loc[:, "Regime"].iloc[-1] = regime_orig
yesbank.tail()
```

```python
# Create a DataFrame with trades, including the price at the trade and the regime under which the trade is made.
yesbank_signals = pd.concat([
        pd.DataFrame({"Price": yesbank.loc[yesbank["Signal"] == 1, "Adj Close"],
                     "Regime": yesbank.loc[yesbank["Signal"] == 1, "Regime"],
                     "Signal": "Buy"}),
        pd.DataFrame({"Price": yesbank.loc[yesbank["Signal"] == -1, "Adj Close"],
                     "Regime": yesbank.loc[yesbank["Signal"] == -1, "Regime"],
                     "Signal": "Sell"}),
    ])
yesbank_signals.sort_index(inplace = True)
yesbank_signals
```
